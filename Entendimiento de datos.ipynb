## 1. Carga y manipulación de los Datos

Para este notebook se usará un set de datos para cada uno de los 4 años que se quieren analizar (2017, 2018, 2021, 2022)

```python
# Importaciones de bibliotecas
!pip  install xlrd pandas numpy
import pandas as pd
import numpy as np
```

```python
nac2017 = pd.read_csv("nac2017.csv", delimiter=",", encoding="latin1")
nac2018 = pd.read_csv("nac2018.csv", delimiter=",", encoding="latin1")
nac2021 = pd.read_csv("nac2021.csv", delimiter=",", encoding="latin1")
nac2022 = pd.read_csv("nac2022.csv", delimiter=",", encoding="latin1")

print(f"Las dimensiones de nac2017 son: {nac2017.shape}")
print(f"Las dimensiones de nac2018 son: {nac2018.shape}")
print(f"Las dimensiones de nac2021 son: {nac2021.shape}")
print(f"Las dimensiones de nac2022 son: {nac2022.shape}")
```

```python
variablesABuscar = {
    # Caracterización de la fecha
    "ANO",  # Año de ocurrencia del nacimiento
    "MES",  # Mes de la ocurrencia del nacimiento
    
    # Caracterización de las regiones específicas
    "COD_DPTO",  # Departamento de nacimiento
    "COD_MUNIC",  # Municipio de nacimiento
    "AREANAC",  # Área de nacimiento (Cabecera municipal, centro poblado, rural disperso)
    
    # Factores socioeconómicos de la madre
    "CODPTORE",  # Departamento de residencia habitual de la madre
    "CODMUNRE",  # Municipio de residencia habitual de la madre
    "AREA_RES",  # Área de residencia habitual de la madre
    "EDAD_MADRE",  # Edad de la madre a la fecha del parto (Rangos de edad)
    "EST_CIVM",  # Estado conyugal de la madre
    "NIV_EDUM",  # Último nivel de estudio que aprobó la madre
    "ULTCURMAD",  # Último año o grado aprobado de la madre
    "SEG_SOCIAL",  # Régimen de seguridad social en salud de la madre
    "NUMCONSUL",  # Número de consultas prenatales que tuvo la madre del nacido vivo
    
    # Caracterización clínica de la madre
    "N_HIJOSV",  # Número de hijos nacidos vivos que ha tenido la madre, incluido el presente
    "N_EMB",  # Número de embarazos, incluido el presente
    
    # Caracterización del nacido vivo
    "SEXO",  # Sexo del nacido vivo
    "PESO_NAC",  # Peso del nacido vivo, al nacer
    "TALLA_NAC",  # Talla del nacido vivo, al nacer
    "T_GES",  # Tiempo de gestación del nacido vivo (Menos de 22, de 22 a 27, de 28 a 37, de 38 a 41, de 42 y más)
    "T_GES_AGRU_CIE",  # Tiempo de gestación del nacido vivo ajustado a la agrupación sugerida por la CIE
    
    # Caracterización del parto
    "TIPO_PARTO",  # Tipo de parto de este nacimiento
    "MUL_PARTO",  # Multiplicidad del embarazo (Simple, doble, triple, cuádruple o más)
    
    # Factores socioeconómicos del padre
    "EDAD_PADRE",  # Edad del padre a la fecha del parto
    "NIV_EDUP",  # Nivel educativo del padre, último año de estudio que aprobó el padre
    "ULTCURPAD",  # Último año o grado aprobado del padre
    
    # Caracterización de la atención del parto
    "SIT_PARTO",  # Sitio de parto (Institución de salud, domicilio)
    "ATEN_PAR",  # El parto fue atendido por
    "PROFESION"  # Profesión de quien certifica el nacimiento
}


columnasNac2017 = [col for col in nac2017.columns if col in variablesABuscar]
columnasNac2018 = [col for col in nac2018.columns if col in variablesABuscar]
columnasNac2021 = [col for col in nac2021.columns if col in variablesABuscar]
columnasNac2022 = [col for col in nac2022.columns if col in variablesABuscar]

columnasNac2017.sort()
columnasNac2018.sort()
columnasNac2021.sort()
columnasNac2022.sort()

print(f"Las columnas de  para los nacimientos en 2017 son: {columnasNac2017}")
print(f"Las columnas de  para los nacimientos en 2018 son: {columnasNac2018}")
print(f"Las columnas de  para los nacimientos en 2021 son: {columnasNac2021}")
print(f"Las columnas de  para los nacimientos en 2022 son: {columnasNac2022}")

columnasEnComun = set(columnasNac2017).intersection(set(columnasNac2018)).intersection(set(columnasNac2021)).intersection(set(columnasNac2022))

print(f"Columnas en común: {columnasEnComun}")
print(f"columnas faltantes en algún año: {variablesABuscar - columnasEnComun}")
```

Se puede evidenciar que todos los sets de datos están completos en términos de las columnas, es decir, todos los años se tienen las mismas variables de interés.

Una vez analizamos los datos por separado procedemos a unir todo en un solo DataFrame.

```python
df_nacimientos = pd.concat([nac2017[columnasNac2017], nac2018[columnasNac2018], nac2021[columnasNac2021], nac2022[columnasNac2022]], axis=0)
df_nacimientos.reset_index(drop=True, inplace=True)
df_nacimientos.sample(30)
```

## Completitud
Se realiza en analisis para ver la completitud de los datos.

```python
porcentajeIncompletitudAtributo =  (100 * df_nacimientos.isnull().sum() / df_nacimientos.shape[0]).sort_values(ascending=False)
pd.DataFrame({
    "Columna": porcentajeIncompletitudAtributo.index,
    "Completitud (%)": 100 - porcentajeIncompletitudAtributo.values,
})
```

El atributo más incompleto es CODMUNRE, con apenas el 0.82% de datos faltantes. Es por esto por lo que podemos decir que si hay completitud en los datos.

## Análisis de unicidad
Se realiza el análisis de unicidad para verificar si hay duplicados.

```python
duplicados = df_nacimientos.duplicated()
total_duplicados = duplicados.sum()
total_datos = df_nacimientos.shape[0]
proporcion_duplicados = 100.0 * total_duplicados / total_datos
print(f"Se encontraron  {total_duplicados} registros duplicados equivalentes al {proporcion_duplicados:.2f}% de los datos")

```

Ya que el porcentaje de duplicados es tan bajo, resulta razonable eliminarlos. 

## Validez








```python
for columna in df_nacimientos:
    print(columna, ":", df_nacimientos[columna].unique())
```

Todas las columnas tienen valores validos de acuerdo con el diccionario de los datos proporcionado por el DANE. Sin mebargo, las columnas ULTCURMAD y ULTCURPAD tienen el valor 0 cuando estas métricas no se encuentran dentro de los posibles valores que pueden tomar.

## Consistencia



Existen varios problemas de consistencia al tratar a los valores nulos. A continuación se iran mencionando y corrigiendo

- Se reemplazan valores inválidos por el valor correspondiente a "Sin información" en las variables `ULTCURMAD` y `ULTCURPAD`

```python
df_nacimientos['ULTCURMAD'] = df_nacimientos['ULTCURMAD'].replace(0, 99)
df_nacimientos['ULTCURPAD'] = df_nacimientos['ULTCURPAD'].replace(0, 99)
```

- Existen valores nulos, pese a que existen atributos para denotar carencia de información

```python
df_nacimientos['PROFESION'] = df_nacimientos['PROFESION'].fillna(9).astype(int)
df_nacimientos['AREA_RES'] = df_nacimientos['AREA_RES'].fillna(9).astype(int)
```

- Las variables `CODMUNRE`y `CODPTORE` no cuentan con un valor para "Sin información". Debido a que son pocas las filas que presentan valores nulos, se procederá a eliminarlas.

```python
prev_len = df_nacimientos.shape[0]
df_nacimientos = df_nacimientos.dropna()
print(f"Se eliminaron {prev_len - df_nacimientos.shape[0]} registros")
```

Finalmente, se exporta el archivo corregido

```python
df_nacimientos.to_csv("NacCompleto.csv", index=False)
```
